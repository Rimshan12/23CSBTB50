```python
import pandas as pd

# Data
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Creating the DataFrame
healthcare_df = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})

# Display the DataFrame
print("Health Care Dataset:")
print(healthcare_df)
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Age vs. Height
plt.figure(figsize=(8, 6))
plt.scatter(healthcare_df['Age'], healthcare_df['Height'], color='blue', marker='o')

plt.title('Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Bar chart: Age vs. Height and Weight
width = 0.35  # width of the bars
fig, ax = plt.subplots(figsize=(10, 6))

# Positions of bars on x-axis
index = np.arange(len(healthcare_df['Age']))

# Bars
bar1 = ax.bar(index - width/2, healthcare_df['Height'], width, label='Height (cm)', color='skyblue')
bar2 = ax.bar(index + width/2, healthcare_df['Weight'], width, label='Weight (kg)', color='salmon')

# Adding labels and title
ax.set_xlabel('Age')
ax.set_ylabel('Height / Weight')
ax.set_title('Height and Weight by Age')
ax.set_xticks(index)
ax.set_xticklabels(healthcare_df['Age'])
ax.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(healthcare_df['Height'], bins=5, color='lightgreen', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')

# Histogram for Weight
plt.subplot(1, 2, 2)
plt.hist(healthcare_df['Weight'], bins=5, color='lightcoral', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
